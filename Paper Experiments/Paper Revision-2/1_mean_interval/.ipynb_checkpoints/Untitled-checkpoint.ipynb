```python
import seaborn as sns
sns.set_theme(style='darkgrid')
import pandas as pd
```

```python
df = pd.DataFrame()

for i in range(10):
    df_tmp = pd.read_csv('data_discrete_{}.csv'.format(i), names=['average conflict number per two flight hours'])
    df_tmp['episode'] = df_tmp.index
    df = pd.concat([df, df_tmp])
df['average conflict number per two flight hours'] = -df['average conflict number per two flight hours']
```

```python
sns.lineplot(x='episode', y='average conflict number per two flight hours', data=df)
```
